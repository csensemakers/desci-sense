```python
# https://pypi.org/project/nanopub/

# trying to change the default `NANOPUB_TEST_SERVER` from [`http://test-server.nanopubs.lod.labs.vu.nl/`](http://test-server.nanopubs.lod.labs.vu.nl/)  to `'http://app.tkuhn.eculture.labs.vu.nl/nanopub-server-1/'`

from pathlib import Path
import rdflib
from rdflib import Graph
from nanopub import Nanopub, NanopubConf, load_profile

```

```python
# 1. Create the config
np_conf = NanopubConf(
    use_test_server=True,
    profile=load_profile(), # Loads the user profile that was created with `np setup`
    add_prov_generated_time=True,
    attribute_publication_to_profile=True,
)
```

```python
my_assertion = Graph()
my_assertion.add((
    rdflib.URIRef('www.example.org/timbernerslee'),
    rdflib.RDF.type,
    rdflib.FOAF.Person
))
```

```python
# 2. Make a Nanopub object with this assertion
np = Nanopub(
    conf=np_conf,
    assertion=my_assertion
)
```

```python
# 3. Publish the Nanopub object
np.publish()
print(np)
```
